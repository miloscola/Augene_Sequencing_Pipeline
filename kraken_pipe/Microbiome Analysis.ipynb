```python
%%bash
#determine microbe ratios with bracken
for file_1 in kraken_reports/*.report; do
    base="${file_1#kraken_reports/}"
    base="${base%.report}"
    bracken -d my_kraken_db/krakendb -i "${file_1}" -r 100 -l S -t 10 \
    -o bracken_outputs/${base}.bracken -w breports/${base}.breport
done

#bracken -d my_kraken_db/krakendb -i kraken_reports/lib3.report -r 100 -l S -t 10 \
#    -o bracken_outputs/lib3.bracken -w breports/lib3.breport
```
