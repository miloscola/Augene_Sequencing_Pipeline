```python
import pandas as pd
import numpy as np
import os
```

```python
def process_bracken_file(file_path):
    # read bracken file
    data = pd.read_csv(file_path, sep='\t', header=0)
    species_data = data[data['taxonomy_lvl'] == 'S'][['name', 'fraction_total_reads']]

    # rename columns
    species_data.columns = ['species', 'relative_abundance']

    # creating a numpy array with relative abundances
    abundance_array = species_data['relative_abundance'].values

    return species_data, abundance_array

# bracken file paths
bracken_file_paths = [
    'kraken_data/SRR14092160.bracken',
    'kraken_data/SRR14092310.bracken',
    'kraken_data/SRR14143424.bracken'
]

# processing all bracken files
for file_path in bracken_file_paths:
    species_abundance_df, abundance_array = process_bracken_file(file_path)
    
    # displaying df and numpy array
    print(f"Data from {file_path}:")
    print(species_abundance_df)
    print(abundance_array)
    print("\n")
```
