```python
import os
import pandas as pd
import subprocess
import gzip
import shutil
import os

def compress_file(filename):
    # Check if the file exists
    if not os.path.exists(filename):
        print("Error: File does not exist.")
        return
    
    # Define the name of the compressed file
    compressed_filename = filename + '.gz'

    # Open the original file in binary read mode
    with open(filename, 'rb') as f_in:
        # Open the compressed file in binary write mode
        with gzip.open(compressed_filename, 'wb') as f_out:
            # Copy the content of the original file to the compressed file
            shutil.copyfileobj(f_in, f_out)

    # Remove the original file
    os.remove(filename)

    print(f"File '{filename}' has been compressed to '{compressed_filename}'.")


# Read the Excel file
df = pd.read_csv('DataFrame_2_Pruned.csv', low_memory=False)

# Create a folder to store the FASTQ files if it doesn't exist
output_folder = 'seq_reads'
os.makedirs(output_folder, exist_ok=True)

# Extract values from the "run" column
runs = df['Run']

# Iterate over each run and run fastq-dump command
for run in runs:
    subprocess.run(['fastq-dump', 'split-files', run])
    
    # Move the generated FASTQ files to the output folder
    for filename in os.listdir('.'):
        if filename.endswith('.fastq'):
            compress_file(os.path.join(filename))
    for filename in os.listdir('.'):
        if filename.endswith('.gz'):        
            os.rename(filename, os.path.join(output_folder, filename))
            
#TODO: rerun on subsetted dataset
```
